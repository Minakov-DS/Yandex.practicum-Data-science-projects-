# Исследование объявлений о продаже квартир

**Цель работы:** исследовать факторы, влияющие на цену продажи квартиры.

**Исходные данные:** архив объявлений о продаже квартир в Санкт-Петербурге за несколько лет от сервиса Яндекс.Недвижимость.

Результаты исследования планируется использовать при создании автоматизированной системы, отслеживающей аномалии и мошенническую деятельность при продаже квартир.

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**Описание данных:**

* airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
* balcony — число балконов
* ceiling_height — высота потолков (м)
* cityCenters_nearest — расстояние до центра города (м)
* days_exposition — сколько дней было размещено объявление (от публикации до снятия)
* first_day_exposition — дата публикации
* floor — этаж
* floors_total — всего этажей в доме
* is_apartment — апартаменты (булев тип)
* kitchen_area — площадь кухни в квадратных метрах (м²)
* last_price — цена на момент снятия с публикации
* living_area — жилая площадь в квадратных метрах(м²)
* locality_name — название населённого пункта
* open_plan — свободная планировка (булев тип)
* parks_around3000 — число парков в радиусе 3 км
* parks_nearest — расстояние до ближайшего парка (м)
* ponds_around3000 — число водоёмов в радиусе 3 км
* ponds_nearest — расстояние до ближайшего водоёма (м)
* rooms — число комнат
* studio — квартира-студия (булев тип)
* total_area — площадь квартиры в квадратных метрах (м²)
* total_images — число фотографий квартиры в объявлении

В случае отсутствия такой замены, при условии, что это в целом не мешает выполняемому анализу данных, оставим данные пропуски без изменений.

Рассмотрим подробнее столбец "is_apartment":

```python
print(df["is_apartment"].value_counts())
```

Подавляющая часть квартир не является апартаментами. Относительно пропусков можно предположить, что в случае если квартира не является апартаментами, то соответствующая ей ячейка в столбце "is_apartment" не заполнялась. Заполним все пропуски значениями False. Данный подход в нашем случае является обоснованным вследствие маленькой относительной доли апартаментов, а также того, что рассматриваемый параметр (принадлежность к апартаментам) непосредственным образом не используется в настоящем исследовании.

Заполнение пропусков в столбце "is_apartment" привело также к изменению типа данных в нем с object на bool.

Рассмотрим подробнее столбец "balcony_number":

```python
print(df["balcony_number"].value_counts())
```

Логично предположить, что пропускам в столбце "balcony_number" соответствуют квартиры без балкона. Таким образом, необходимо заменить пропуски в рассматриваемом столбце нулями:

```python
df["balcony_number"] = df["balcony_number"].fillna(0)
```

С учетом того, что количество балконов не может быть дробным числом, выполним замену типа данных для столбца "balcony_number":

```python
df["balcony_number"] = df["balcony_number"].astype("int")
```

В столбце "locality_name" содержится весьма значительное количество уникальных населенных пунктов. При этом существует вероятность наличия повторов из-за отличий в написании одного и того же населенного пункта. Для исключения данной вероятности необходим подробный и трудоемкий анализ с применением в том числе метода лемматизации. В рамках нашего исследования ограничимся исключением вероятности повторов в названиях чаще всего встречаемых в датафрейме населенных пунктов. Для этого выведем названия 20 наиболее часто встречаемых населенных пункта:

При этом имеется повтор: "кудрово", "деревня кудрово". "деревня кудрово" - является устаревшим названием населенного пункта. Выполним замену "деревня кудрово" на "кудрово":

Замену пропусков для столбца "city_center_distance" логично выполнить медианными значениями, рассчитанными для каждого конкретного населенного пункта. Составим таблицу медианных значений расстояния до центра города в зависимости от населенного пункта и выведем ее первые пять строк. 

```python
medians_city_center_distance = df.groupby("locality_name")["city_center_distance"].median()
medians_city_center_distance.head()
```

Все первые пять строк содержат NaN. Подсчитаем долю пропусков в данной таблицы относительно всех значений:

```python
print(medians_city_center_distance.isnull().sum() / len(medians_city_center_distance))
```

92% - очень высокая доля пропусков. Очевидно, что заполнить существенное количество пропусков с использованием данной таблицы неудастся. Тем не менее выполним замену:

```python
for locality_name in medians_city_center_distance.index:
    mask = df["locality_name"] == locality_name
    df.loc[mask, "city_center_distance"] = df.loc[mask, "city_center_distance"].fillna(
        medians_city_center_distance[locality_name])
```

Подсчитаем количество пропусков в столбце "city_center_distance" после замены:

```python
print(df["city_center_distance"].isnull().sum())
```

Количество пропусков немного уменьшилось: с 5519 до 5394. Выполним аналогичные действия для столбца "airport_distance":

```python
# cоставим таблицу медианных значений расстояния до аэропорта в зависимости от населенного пункта
medians_airport_distance = df.groupby("locality_name")["airport_distance"].median()
# заполним пропуски с использованием медианных значений
for locality_name in medians_airport_distance.index:
    mask = df["locality_name"] == locality_name
    df.loc[mask, "airport_distance"] = df.loc[mask, "airport_distance"].fillna(
        medians_airport_distance[locality_name])
# подсчитаем количество пропусков в столбце "airport_distance" после замены
print(df["airport_distance"].isnull().sum())
```

## Рассчёты и добавление результатов в таблицу

Подсчитаем и добавим в таблицу следующие данные: 
* цену квадратного метра 
* день недели, месяц и год публикации
* этаж квартиры (три варианта: - первый, последний, другой)
* отношение жилой площади к общей площади
* отношение площади кухни к общей площади

```python
# добавляем в таблицу новый столбец "square_meter_price" с указанием цены квадратного метра
df["square_meter_price"] = df["last_price"] / df["total_area"]
```

```python
# добавляем в таблицу новые столбцы с указанием дня недели, месяца и года
df["weekday_exposition"] = df["first_day_exposition"].dt.weekday
df["month_exposition"] = df["first_day_exposition"].dt.month
df["year_exposition"] = df["first_day_exposition"].dt.year
```

```python
# программа для категоризации этажа квартиры на три категории: "первый", "последний" и "другой" 
def floor_group(row):
    if row[6] == 1:
        return "первый"
    if row[8] == 1:
        return "первый"
    if row[8] == row[6]:
        return "последний"
    else:
        return "другой"
```

```python
# добавляем в таблицу столбец с результатами категоризации по этажу, используя программу floor_group
df["floor_group"] = df.apply(floor_group, axis=1)
```

```python
# добавляем в таблицу столбец с отношением жилой площади к общей площади
df["living_to_total_area_ratio"] = df["living_area"] / df["total_area"]
```

```python
# добавляем в таблицу столбец с отношением площади кухни к общей площади
df["kitchen_to_total_area_ratio"] = df["kitchen_area"] / df["total_area"]
```

### Исследование параметров

**Изучим пять параметров: полная площадь, цена, число комнат, высота потолков, время продажи квартиры**.  
Изучение параметров будем проводить с использованием следующих инструментов: гистограмм, "ящика с усами", а также метода describe(). Для каждого из параметров определим редкие, выбивающиеся значения, и диапазон, полученный исключением данных значений. Сравним влияние каждого из параметров на стоимость квартиры.

Для определения нижнего и верхнего "усов" напишем программу "whiskers" (пер. с англ.: "усы"):

```python
def whiskers(columns):
    q_1 = columns.quantile(0.25)
    q_3 = columns.quantile (0.75)
    interquartile_range = q_3 - q_1
    lower_whisker = q_1 - 1.5*interquartile_range
    upper_whisker = q_3 + 1.5*interquartile_range
    if lower_whisker < columns.min():
        lower_whisker = columns.min()
    if upper_whisker > columns.max():
        upper_whisker = columns.max()
    return print("Нижний ус: {:.1f}.\nВерхний ус: {:.1f}.".format(lower_whisker, upper_whisker))
```

### Исследование факторов, влияющих на цену продажи

**Изучим какие факторы больше всего влияют на стоимость квартиры**  

Рассмотрим влияние следующих факторов на стоимость квартиры: площади, числа комнат, удаленности от центра, категории этажа (первый, последний, другой) и даты размещения: дня недели, месяца и года. Для оценки влияния на стоимость квартиры мы имеем на текущий момент два основных инструмента: визуальный при помощи графиков и определение коэффициента линейной корреляции Пирсона. Графический (визуальный) метод универсален, однако он субъективен (зависит от специалиста, оценивающего графики). Использование коэффициента Пирсона обосновано не для всех факторов. Из рассматриваемых нами факторов он хорошо подходит для площади, числа комнат и удаленности от центра. Для оценки влияния на цену категории этажа и даты размещения он подходит плохо. Это связано с тем, что в случае этажей мы имеем категориальную переменную (три категории: первый,последний, другой), а не числовую, а в случае даты размещения мы имеем скорее всего нелинейную зависимость, а циклическую или какую-то более сложную. Всвязи с этим для факторов категория этажа и даты размещения ограничимся анализом на основе графиков и сводных таблиц.

### Определение границ центра Санкт-Петербурга

**Определим границы центра Санкт-Петербурга**

```python
# выделим данные для Санкт-Петербурга.
df_peterburg = pd.DataFrame(df.query("locality_name == 'санкт-петербург'"))

# **********************для ревьюера************************
# здесь срез методом query() оборачивается методом pd.DataFrame() вследствие того,
# что в противном случае при создании нового столбца следующим способом:  
# df_peterburg["center_distance_km"] = df_peterburg["city_center_distance"] / 1000 получается следующая ошибка:
# "A value is trying to be set on a copy of a slice from a DataFrame". Почему метод pd.DataFrame() позволяет ее обходить
# не совсем ясно, так как при проверке методом type() между df.query("locality_name == 'санкт-петербург'") и
# pd.DataFrame(df.query("locality_name == 'санкт-петербург'")) нет разницы (class 'pandas.core.frame.DataFrame').
# По непонятной причине срез методом query() не является полноценным объектом DataFrame...
#***********************************************************

# создаем новый столбец "center_distance_km" с расстояниями до центра (целые числа)
df_peterburg["center_distance_km"] = (df_peterburg["city_center_distance"] / 1000).round(0).astype("int")
# выведем для контроля первые пять строк таблицы df_peterburg со столбцами: "locality_name", "city_center_distance", "center_distance_km"
df_peterburg[["locality_name", "city_center_distance", "center_distance_km"]].head()
print(len(df_peterburg))
```

```python
# построим сводную таблицу для определения средней цены квартиры в зависимости от расстояния в км от центра Санкт-Петербурга
df_peterburg.pivot_table(index="center_distance_km", values="last_price")
```

При построении сводной таблицы сознательно использовались неочищенные данные по last_price. Это связано с тем, что при выбраковке данных нами исключались из рассмотрения в основном очень дорогие квартиры. Данные квартиры расположены главным образом в центре, и их исключение из выборки, вероятно, очень сильно исказит картину именно для центра города, границы которого мы исследуем.

```python
# визуализируем полученные результаты при помощи графика
df_peterburg.pivot_table(index="center_distance_km", values="last_price").plot(style='o-', figsize=(10, 5), grid=True)
plt.show()
```

Рассмотрим полученный график. Мы имеем довольно сложную зависимость. Сперва до отметки 3 км цены очень быстро падают по мере удаления от центра города, затем снова возрастают до отметки 6-7 км, затем они снова резко падают. Данный график, вероятно, говорит о том, что простое деление города на две части: центр / окраины не совсем верно отражает реальную зависимость. Явно выделяется обособленная зона в диапазоне от 3 до 7 км, где по каким-то причинам сложились условия для высоких цен на квартиры. В данной зоне, к примеру, на расстояни 7 км цены сущетвенно больше, чем на расстоянииии 3 км. Приниципиально на основании полученного графика мы можем выделить размер центра различными путями. Условно примем размер (радиус) центра равным 7 км, исходя из того, что именно после 7 км начинается резкое падение цен без последующих их существенных подъемов.
